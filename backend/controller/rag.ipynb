```python
#apis

pinecone_api_key =os.getenv("PINECONE_API_KEY")
groq_api_key = os.getenv("GROQ_API_KEY")
llama_cloud_api_key = os.getenv("LLAMA_CLOUD_API_KEY")


# All set till here
print(pinecone_api_key)
print(groq_api_key)
print(llama_cloud_api_key)
```

```python
from pinecone import Pinecone, ServerlessSpec
import time

# Initialize Pinecone 
pc = Pinecone(api_key=pinecone_api_key)

# Define the index name
index_name = "test-databasee"  

existing_indexes = pc.list_indexes()
   
if index_name not in existing_indexes:
    try:
        pc.create_index(
            name=index_name,
            dimension=384,  
            metric='cosine',
            spec=ServerlessSpec(
                cloud='aws',
                region='us-east-1'
            ),
            deletion_protection='enabled'  # Enable delete protection
        )
        time.sleep(20)
        
        
        index = pc.Index(index_name)
        print(f"Successfully created and connected to index: {index_name}")
        
    except Exception as e:
        print(f"Error creating/connecting to index: {str(e)}")
```

```python
#load the pdf /scan pdf etc

import pymupdf 
doc = pymupdf.open(r"C:\Users\HP\Desktop\Project\artificizen\backend\129049121_1734976290923.pdf") 
for page_num in range(doc.page_count):
    page = doc.load_page(page_num)  # Load the page by its number
    text = page.get_text()
    print(text)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


def text_splitter(text):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=200
    )
    return text_splitter.split_text(text)  



data = [] #return data after preprocessing will store in the data list
for page_num in range(doc.page_count):
    page = doc.load_page(page_num)
    text = page.get_text()
    
    chunks = text_splitter(text)  
    data.extend(chunks)  

print("Text has been converted into chunks.")
```

```python
#embeddings
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name = "sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs = {"device": "cpu"}
)

embeddings.embed_documents("Hello i am calling from function")
```

```python
from langchain.vectorstores import Pinecone

# Convert your text chunks into Document objects
document = [Document(page_content=chunk) for chunk in data]
vectorstore = Pinecone.from_documents(documents=document,
                                    embedding=embeddings,
                                    index_name=index_name)
```

```python
retriver =  vectorstore.as_retriever(search_kwargs={"k": 1})
```

```python
response=retriver.invoke("What is this pdf ?")
print(response)
```

```python
#setup the models
Model = ChatGroq(
    model='deepseek-r1-distill-llama-70b',
    api_key = groq_api_key,
    verbose=True
)
```

```python
from langchain.chains import RetrievalQA
qa_chain = RetrievalQA.from_chain_type(
    llm=Model,
    chain_type="stuff",
    retriever=retriver,
    return_source_documents=True,
)
```

```python
result=qa_chain.invoke("Bring the all transaction number from the pdf?")

# Get the answer from the result dictionary
answer = result['result']

# Print the answer with line breaks for better readability
print(answer.replace('. ', '.\n'))
```
